```python
import os
import sys
import joblib
import pickle
import numpy as np
import pandas as pd

from pandas import read_csv
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from imblearn.over_sampling import SMOTE
import imblearn


sys.path.append(os.path.abspath("ColumnsProcessing.py"))
from ColumnsProcessing import *
```

```python
cp = ColumnsProcessing()
pd.set_option("display.max_columns", 100)
```

```python
fileName = 'facts_training_dataset_1617463082875'

# Load dataset
# Este archivo balanceado es para pruebas.
# Igual se deben arreglar campos en el dataset de entrada para acomodar las entradas de tipo -O)-
df_ = read_csv('/Users/borisperezg/rebelmodels_storing/datasets_to_train/'+fileName+'_pa.csv')
```

```python
df_
```

```python
df_.actiontype.value_counts()
```

```python
# --------------------------------------
# CORRECCION DE NULLS
# --------------------------------------
```

```python
df_[['targetelementname']] = df_[['targetelementname']].fillna(value='NoName')
df_[['targetelementtype']] = df_[['targetelementtype']].fillna(value='NoType')
df_[['sourceelementtype']] = df_[['sourceelementtype']].fillna(value='NoType')
df_[['mostlinkedlayer']] = df_[['mostlinkedlayer']].fillna(value='NoLayer')
df_[['isnewelement']] = df_[['isnewelement']].fillna(value='NoNew')
df_[['isnewelement']] = df_[['isnewelement']].astype(str)

df_[['layertarget']] = df_[['layertarget']].fillna(value='NoLayer')
df_[['layersource']] = df_[['layersource']].fillna(value='NoLayer')
df_[['relatontype']] = df_[['relatontype']].fillna(value='NoRelation')

df_[['property']] = df_[['property']].fillna(value='NoProperty')
df_[['propertynewvalue']] = df_[['propertynewvalue']].fillna(value='NoValue')
df_[['propertyoldvalue']] = df_[['propertyoldvalue']].fillna(value='NoValue')

#df_[['commitlogs_ngrams']] = df_[['commitlogs_ngrams']].fillna(value='NoData')
#df_[['chatlogs_ngrams']] = df_[['chatlogs_ngrams']].fillna(value='NoData')
#df_[['adrlogs_ngrams']] = df_[['adrlogs_ngrams']].fillna(value='NoData')
```

```python
# --------------------------------------
# LEMATIZACION Y REMOCION DE STOPWORDS
# --------------------------------------
```

```python
df_['driver_lemma'] = df_.drivertype.apply(cp.remove_stopwords_and_lemma)
df_['goal_lemma'] = df_.goaltype.apply(cp.remove_stopwords_and_lemma)
df_['sourceelementname_lemma'] = df_.sourceelementname.apply(cp.remove_stopwords_and_lemma)
df_['targetelementname_lemma'] = df_.targetelementname.apply(cp.remove_stopwords_and_lemma)

# Los campos commitlogs_ngrams, chatlogs_ngrams, adrlogs_ngrams ya fueron procesados en Step 1
```

```python
# --------------------------------------
# CONCATENACION DE CAMPOS TEXTUALES LEMATIZADOS
# --------------------------------------
```

```python
df_['all_texts'] = df_['driver_lemma'].map(str) + ' ' + df_['goal_lemma'].map(str) + ' ' +  df_['sourceelementname_lemma'].map(str) + ' ' +  df_['targetelementname_lemma'].map(str) + ' ' +  df_['commitlogs_ngrams'].map(str) + ' ' +  df_['chatlogs_ngrams'].map(str) + ' ' +  df_['adrlogs_ngrams'].map(str)
```

```python
df_.drop(['drivertype', 'goaltype', 'sourceelementname', 'driver_lemma', 'goal_lemma', 
          'sourceelementname_lemma', 'targetelementname', 'targetelementname_lemma',
         'commitlogs_ngrams', 'chatlogs_ngrams', 'adrlogs_ngrams'], axis=1, inplace=True)
```

```python
# --------------------------------------
# VECTORIZACION DE COLUMNA TEXTUAL
# --------------------------------------
```

```python
df_.actiontype.value_counts()
```

```python
result = cp.textColumnsVectorization(df_, 'all_texts')
```

```python
result
```

```python
# --------------------------------------
# NORMALIZACION DE COLUMNAS NUMERICAS
# --------------------------------------
```

```python
result = cp.numericalNormalization(result, 'incoming')
```

```python
result = cp.numericalNormalization(result, 'outcoming')
```

```python
# --------------------------------------
# PROCESAMIENTO DE COLUMNAS CATEGORICAS
# --------------------------------------
```

```python
result = cp.openCategoricalColumns(result, 'sourceelementtype')
result = cp.openCategoricalColumns(result, 'targetelementtype')
result = cp.openCategoricalColumns(result, 'propertynewvalue')
result = cp.openCategoricalColumns(result, 'propertyoldvalue')
result = cp.openCategoricalColumns(result, 'mostlinkedlayer')
result = cp.openCategoricalColumns(result, 'isnewelement')
result = cp.openCategoricalColumns(result, 'layersource')
result = cp.openCategoricalColumns(result, 'layertarget')
result = cp.openCategoricalColumns(result, 'relatontype')
result = cp.openCategoricalColumns(result, 'actiontype')
result = cp.openCategoricalColumns(result, 'property')
result = cp.openCategoricalColumns(result, 'facttype')
```

```python
# result.to_csv('/Users/borisperezg/rebelmodels_storing/datasets_to_train/result.csv', index=False, index_label=True)
```

## ---------------------------------------------
## ENTRENAMIENTO PARA ATD

```python
# --------------------------------------------------------------------
# PREPARACION DE LOS DATOS PARA ENTRENAR LA IDENTIFICACION DE ATD
# --------------------------------------------------------------------
```

```python
XA = result.drop(['atdcause','affectedqa'], axis=1)
yA = result['atdcause']
XA_train, XA_test, yA_train, yA_test = train_test_split(XA, yA, test_size=0.2, random_state=1, shuffle=True)
```

```python
# --------------------------------------------------------------------
# BALANCEANDO EL CLASS EN CASO DE NECESITARSE
# --------------------------------------------------------------------
```

```python
df_concat = pd.concat([XA_train, yA_train], axis=1, ignore_index=True)
df_concat
```

```python

#sm = SMOTE(random_state=27, ratio=1.0)
sm = SMOTE()
```

```python
try:
    XA_train, yA_train = sm.fit_resample(XA_train, yA_train)
except Exception as e:
    # Se hace upsampling de la clase minoritaria
    print('Error by SMOTE, manual upsampling')
    # concatenate our training data back together
    df_concat = pd.concat([XA_train, yA_train], axis=1, ignore_index=True)
    df_concat.rename(columns={123: 'atdcause'}, inplace=True)
    is_atd = df_concat['atdcause'] != 'noatd'
    df_try = df_concat[is_atd]
    newdf = pd.DataFrame(np.repeat(df_try.values, 3, axis=0), columns=df_try.columns)
    df_final = df_concat.append(newdf, ignore_index=True)
    XA_train = df_final.drop(['atdcause'], axis=1)
    yA_train = df_final['atdcause']
```

```python
#df_concat
yA_train.value_counts()
```

```python
yA_test
```

```python
yA_train
```

```python
# --------------------------------------
# PROBANDO VARIOS ALGORITMOS
# --------------------------------------
```

```python
# Spot Check Algorithms
models = []
models.append(('LR', LogisticRegression(solver='liblinear', multi_class='ovr')))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('CART', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))
models.append(('MNB', MultinomialNB()))
models.append(('SVM', SVC(gamma='auto')))
# evaluate each model in turn
results = []
names = []
for name, model in models:
	kfold = StratifiedKFold(n_splits=7, random_state=1, shuffle=True)
	cv_results = cross_val_score(model, XA_train, yA_train, cv=kfold, scoring='accuracy')
	results.append(cv_results)
	names.append(name)
	print('%s: %f (%f)' % (name, cv_results.mean(), cv_results.std()))
```

```python
# --------------------------------------
# ENTRENAMIENTO DEL MODELO
# --------------------------------------
```

```python
#modelA = MultinomialNB()
#modelA = DecisionTreeClassifier()
#modelA = LinearDiscriminantAnalysis()
modelA = GaussianNB()
modelA.fit(XA_train, yA_train)
```

```python
print(modelA.score(XA_test, yA_test))
```

```python
y_predA = modelA.predict(XA_test)
print(y_predA)
```

```python
try:
    XQ_train, yQ_train = sm.fit_resample(XQ_train, yQ_train)
except Exception as e:
    # Se hace upsampling de la clase minoritaria
    print('Error by SMOTE, manual upsampling')
    # concatenate our training data back together
    
    # Concatena los set de entrenamiento
    df_concat = pd.concat([XQ_train, yQ_train], axis=1, ignore_index=True)
    # ****** Nota: Este indice de columna de affectedqa podria cambiar, 
    # ****** dependiendo del numero de casos de ATD detectados
    # Renombra la ultima columna
    df_concat.rename(columns={127: 'affectedqa'}, inplace=True)
    # Obtiene los registros que tengan un QA afectado
    is_atd = df_concat['affectedqa'] != 'none'
    df_try = df_concat[is_atd]
    # Triplica los registros
    newdf = pd.DataFrame(np.repeat(df_try.values, 3, axis=0), columns=df_try.columns)
    
    df_final = df_concat.append(newdf, ignore_index=True)
    XQ_train = df_final.drop(['affectedqa'], axis=1)
    yQ_train = df_final['affectedqa']
```
